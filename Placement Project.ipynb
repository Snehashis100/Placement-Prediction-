```python
#'mba_p' is having (high outliers)the most number of  outliers,then 'degree_p'(high outlier) and then 'hsc_p' having(low outlier) the least
import numpy as np
def Z_score(feature,df):
    data=df.copy()
    for f in feature:
        m= np.mean(data[f])
        sd=np.std(data[f])
        outliers=[]
        for value in data[f]:
            z_score=(value - m)/sd
            if np.abs(z_score)>3:
                outliers.append(value)
        print(f"\'{f}\' has {outliers} outliers")
Z_score(numerical_features,df)

# According to Z score there are no outliers
```

```python
def iqrr(feature,df):
    data=df.copy()
    for f in feature:
        q1, q3= np.percentile(data[f],[25,75])
        iqr=q3-q1
        outliers=[]
        for  i in data[f]:
            if i<(q1-1.5*(iqr)) or i>(q3+1.5*(iqr)):
                outliers.append(i)
            out_percent=np.round((len(outliers)/300)*100,2)
        print(f"\'{f}\' has {outliers} outliers i.e {out_percent}%" )
iqrr(numerical_features,df)

# But by using the IQR we got the outliers
```

```python
# These outliers must be handeled based on the problem we are solving and the algo which we will be using
# Replacing the outliers with the 100th percentile value and 0th percentile value and making  new columns
def Quantile(data,feature):
    q1, q3= np.percentile(data[feature],[25,75])
    iqr=q3-q1
    lower=q1-1.5*(iqr)
    upper=q3+1.5*(iqr)
    return lower,upper
data=df.copy()
lower,upper=Quantile(data,'hsc_p')
data['hsc_p_out']=np.where(df.hsc_p<lower,int(lower),df.hsc_p)
data['hsc_p_out']=np.where(df.hsc_p>upper,int(upper),df.hsc_p)

lower,upper=Quantile(data,'degree_p')
data['degree_p_out']=np.where(df.degree_p<lower,int(lower),df.degree_p)
data['degree_p_out']=np.where(df.degree_p>upper,int(upper),df.degree_p)

lower,upper=Quantile(data,'mba_p')
data['mba_p_out']=np.where(df.mba_p<lower,int(lower),df.mba_p)
data['mba_p_out']=np.where(df.mba_p>upper,int(upper),df.mba_p)

data['hsc_p_out']=np.where(df.hsc_p==37.0,38,df.hsc_p)

iqrr(['hsc_p_out','mba_p_out','degree_p_out'],data)
```

```python
# converting the marks into grades and making them a new column
def school_to_grade(marks):
    if marks>=90 and marks<=100:
        return 'A'
    elif marks>=80 and marks<=89:
        return 'B'
    elif marks>=66 and marks<=79:
        return 'C'
    elif marks>=40 and marks<=65:
        return 'D'
    else:
        return 'E'

def college_to_grade(marks):
    if marks>=90 and marks<=100:
        return 'A'
    elif marks>=80 and marks<=89:
        return 'B'
    elif marks>=70 and marks<=79:
        return 'C'
    elif marks>=60 and marks<=69:
        return 'D'
    elif marks>=50 and marks<=59:
        return 'E'
    else:
        return 'F'
data['ssc_grade']=data['ssc_p'].apply(school_to_grade)
data['hsc_grade']=data['hsc_p'].apply(school_to_grade)

data['degree_grade']=data['degree_p'].apply(college_to_grade)
data['etest_grade']=data['etest_p'].apply(college_to_grade)
data['mba_grade']=data['mba_p'].apply(college_to_grade)
```

```python
# Making a new column which will be having only 1 when hsc_s and degree_t are same else 0
def Stream(df,feature1,feature2):
    lst=[]
    for i in range(len(df[feature1])):
        if ('Sci' in df.loc[i,feature1] and 'Sci' in df.loc[i,feature2])or ('Com' in df.loc[i,feature1] and 'Com' in df.loc[i,feature2])or('Arts' in df.loc[i,feature1] and 'Others' in df.loc[i,feature2]):
            lst.append(1)
        else:
            lst.append(0)
    return lst

data1['same_stream']=pd.Series(Stream(data1,'hsc_s','degree_t'),dtype='int64')
data1.head(10)
        
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
def Models_Evaluation(X_final,y,classifier):
    accuracy=[]
    skf = StratifiedKFold(n_splits=5,shuffle=True,random_state=1)
    skf.get_n_splits(X_final, y)
    # X is the feature set and y is the target
    for model in classifier:
        accuracy=[]
        for train_index, test_index in skf.split(X_final,y): 
#             print("Train:", train_index, "Validation:", test_index) 
            X1_train, X1_test = X_final.iloc[train_index], X_final.iloc[test_index] 
            y1_train, y1_test = y.iloc[train_index], y.iloc[test_index]

            model.fit(X1_train,y1_train)
            prediction=model.predict(X1_test)
            score=accuracy_score(y1_test,prediction)
            accuracy.append(score)
            mean=sum(accuracy)/5
        print(f"The Accuracies of {model}:{np.round(accuracy,2)} and mean is {np.round(mean,2)}")
```
